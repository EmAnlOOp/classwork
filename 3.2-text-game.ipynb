```python
# create a typed dictionary for charcaters 
from typing import TypedDict, List, Dict

class CharacterTypes (TypedDict): 
    name: str
    health: float 
    attack_power: int
    defense: int 
    strength: str 
    
strengths_weakness = {
    "fire": "water", 
    "water": "earth", 
    "earth": "fire", 
}  

total_points: int = 20 

# default character 
character: CharacterTypes = {
    "name": "Hero", 
    "health": 90.0, 
    "attack_power": 10, 
    "defense": 5, 
    "strength": "water", 
} 
    
# location hashmap 
locations: Dict[str,int] = {
    "forest": 0 
}
    
enemies_list: List[CharacterTypes] = [
    {
        "name" : "Goblin", 
        "health": 20, 
        "attack_power": 5, 
        "defense": 1, 
        "strength": "water", 
    },
]

class GameplayDataTypes(TypedDict): 
    enemies_defeated: List[str]
    damage_taken: List[float]
    damage_dealt: List[float]
    enemies_encountered: List[str]
    
gameplay_data: GameplayDataTypes = {
    "enemies_defeated": [], 
    "damage_dealt": [], 
    "damage_taken": [], 
    "enemies_encountered": [], 
}

from typing import Union
# character build sequence
name = input(f"Choose your character's Name: ")
print(f"""
Welcome {name}.
Let's build your character. 
You have {total_points} to distribute. 
""")
print(f"""
You can distribute points to 
Health, 
attack power, 
defense, 
and choose a strength (fire, water, earth)!
""")

try:
    char_num_options: List[str] = ["health", "attack_power", "defense"]
    for opt in char_num_options: 
        is_complete = False
        while not is_complete: 
            opt_add: Union[str,int] = int(input(f"""
            How many points for {opt}?
            You have {total_points} available: """))
            if opt_add.isnumeric():
                print("Must be only numebrs. Try again")
                continue
            if opt_add < 0 or opt_add > total_points: 
                print(f"Must be between 0 and {total_points}. Try again")
                continue
            opt_add = int(opt_add)
            is_complete = True
        character[opt] += opt_add
        total_points -= opt_add
        print(f"Your character now has {character[opt]} {opt} points")    

except Exception as e: 
    print(f"Error: {e}. Please start over")
    exit(1)
    
print("------Round 1-----")

current_location = input(f"Choose a location to travel to: {', '.join(locations.keys())}")
if current_location not in locations.keys(): 
    print(f"Choose from these options: {', '.join(locations.keys())}")
    print("start over")
    exit(1)
    
enemy1 = enemies_list[locations[current_location]]
print(f"{character['name']} encounters {enemy1['name']}")

choices = ["attack", "wait"]

char_choice1 = input(f"What do you want to do: {' or '.join(choices)}")
if char_choice1 not in choices: 
    print(f"{char_choice1} not valid. Start over")
    exit(1)
if char_choice1 == "attack":
    c1_ap = character["attack_power"] - enemy1 ["defense"]
    enemy1["health"] -= c1_ap if c1_ap > enemy1["defense"] else 0 
    print(f"{character['name']} attacks {enemy1["name"]} for {c1_ap}")
    enemy1_a = enemy1["attack_power"] - character["defense"]
    character["health"] -= enemy1_a if enemy1_a > character["defense"] else 0 
else: 
    pass 
```
